# Outer Joins

```python
import pandas as pd
```

```python
week1 = pd.read_csv("C:\\Users\\Carlos.Villarreal\\Documents\\Python Pandas Course\\pandas\\Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("C:\\Users\\Carlos.Villarreal\\Documents\\Python Pandas Course\\pandas\\Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("C:\\Users\\Carlos.Villarreal\\Documents\\Python Pandas Course\\pandas\\Restaurant - Customers.csv")
foods = pd.read_csv("C:\\Users\\Carlos.Villarreal\\Documents\\Python Pandas Course\\pandas\\Restaurant - Foods.csv")
```

```python
week1.merge(week2, how = "outer", on = "Customer ID")
```

```python
week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - Week 1", " - Week 2"])
```

```python
len(week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - Week 1", " - Week 2"]))
```

```python
week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - Week 1", " - Week 2"], indicator = True)
```

```python
merged = week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - Week 1", " - Week 2"], indicator = True)
```

```python
merged["_merge"].value_counts()
```

```python
merged["_merge"].isin(["left_only", "right_only"])
```

```python
mask = merged["_merge"].isin(["left_only", "right_only"])
```

```python
merged[mask]
```

```python
mask = merged["_merge"].isin(["both"])
```

```python
merged[mask]
```

# Merging by Indexes with the left_index and right_index Parameters

```python
import pandas as pd
```

```python
week1 = pd.read_csv("C:\\Users\\Carlos.Villarreal\\Documents\\Python Pandas Course\\pandas\\Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("C:\\Users\\Carlos.Villarreal\\Documents\\Python Pandas Course\\pandas\\Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("C:\\Users\\Carlos.Villarreal\\Documents\\Python Pandas Course\\pandas\\Restaurant - Customers.csv", index_col = "ID")
foods = pd.read_csv("C:\\Users\\Carlos.Villarreal\\Documents\\Python Pandas Course\\pandas\\Restaurant - Foods.csv", index_col = "Food ID")
```

```python
foods.head(3)
```

```python
customers.head(3)
```

```python
week1.head()
```

```python
week1.merge(customers, how = "left", left_on = "Customer ID", right_index = True)
```

```python
sales = week1.merge(customers, how = "left", left_on = "Customer ID", right_index = True)
sales.merge(foods, how = "left", left_on = "Food ID", right_index = True)
```

```python
week1.merge(week2, how = "left", left_index = True, right_index = True, suffixes = [" - Week 1", " - Week 2"])
```
